```python
import numpy as np
#calcul de TF
tf_t1_d1=1
tf_t4_d1=1
tf_t3_d1=1

tf_t4_d2=1
tf_t5_d2=2
tf_t2_d2=1
tf_t3_d2=1

tf_t2_d3=1
tf_t1_d3=1
tf_t4_d3=2

#Calcul de IDF
nb=3
nb_t1=2
nb_t2=2
nb_t3=2
nb_t4=3
nb_t5=1

idf_t1=np.log10(3/2)
idf_t2=np.log10(3/2)
idf_t3=np.log10(3/2)
idf_t4=np.log10(3/3)
idf_t5=np.log10(3/1)

#calcul de TF_IDF
tf_idf_t1_d1=tf_t1_d1*idf_t1
tf_idf_t1_d2=0
tf_idf_t1_d3=tf_t1_d3*idf_t1

tf_idf_t2_d1=0
tf_idf_t2_d2=tf_t2_d2*idf_t2
tf_idf_t2_d3=tf_t2_d3*idf_t2

tf_idf_t3_d1=tf_t3_d1*idf_t3
tf_idf_t3_d2=tf_t3_d2*idf_t3
tf_idf_t3_d3=0


tf_idf_t4_d1=tf_t4_d1*idf_t4
tf_idf_t4_d2=tf_t4_d2*idf_t4
tf_idf_t4_d3=tf_t4_d3*idf_t4

tf_idf_t5_d1=0
tf_idf_t5_d2=tf_t5_d2*idf_t5
tf_idf_t5_d3=0



mat_ter_doc = [[],[],[],[],[]]
mat_ter_doc[0] = [tf_idf_t1_d1, tf_idf_t1_d2, tf_idf_t1_d3]  # t1
mat_ter_doc[1] = [tf_idf_t2_d1, tf_idf_t2_d2, tf_idf_t2_d3]  # t2
mat_ter_doc[2] = [tf_idf_t3_d1, tf_idf_t3_d2, tf_idf_t3_d3]  # t3
mat_ter_doc[3] = [tf_idf_t4_d1, tf_idf_t4_d2, tf_idf_t4_d3]  # t4
mat_ter_doc[4] = [tf_idf_t5_d1, tf_idf_t5_d2, tf_idf_t5_d3]  # t5

mat_ter_doc = np.array(mat_ter_doc)
print(mat_ter_doc)

#vecteurs de doc
d1 = mat_ter_doc[:,0]
d2 = mat_ter_doc[:,1]
d3 = mat_ter_doc[:,2]

#Similarité cosinus
def cosinus(a,b):
    return np.dot(a,b)/(np.linalg.norm(a)*np.linalg.norm(b))

sim_d1_d2 = cosinus(d1,d2)
sim_d1_d3 = cosinus(d1,d3)

print("cos(d1,d2) =", sim_d1_d2)
print("cos(d1,d3) =", sim_d1_d3)


if sim_d1_d2 > sim_d1_d3:
    print("Le document le plus proche de d1 est d2")
else:
    print("Le document le plus proche de d1 est d3")
```
