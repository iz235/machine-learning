# TP — Classification de texte avec scikit-learn

Dans ce TP, nous construisons un classifieur capable d’identifier la langue
d’une phrase parmi trois langues : **français (fr)**, **anglais (en)** et **italien (it)**.

Principe :
- chaque phrase est transformée en vecteur numérique à partir de la **fréquence des lettres** (caractères) ;
- un modèle de classification est ensuite entraîné et évalué sur des données séparées (train/test).


Récupération du texte (TODO à compléter)

```python
import glob

text = {"fr": "", "en": "", "it": ""}

# On récupère tous les fichiers correspondant à chaque langue
for lang in ["fr", "en", "it"]:
    files = glob.glob(f"data/{lang}_*.txt")

    for path in files:
        with open(path, "r", encoding="utf-8") as f:
            text[lang] += f.read() + "\n"
```

```python
assert "Le Chat grimaça en apercevant Alice" in text["fr"]
print(text["fr"][:200])
```

Étape 2 — Création du tableau (TODO variable data)

```python
from nltk.tokenize import sent_tokenize
import numpy as np

data = []

for lang in ["fr", "en", "it"]:
    sentences = sent_tokenize(text[lang])
    for s in sentences:
        s = s.strip()
        if len(s) > 0:
            data.append([lang, s])

assert len(np.array(data).shape) == 2 and np.array(data).shape > (1000, 2)
data[:5]
```
