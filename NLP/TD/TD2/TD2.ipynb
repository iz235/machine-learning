# Partie A : Préparation du corpus

Etape 1:

corpus est une liste de documents textuels.

Chaque élément = une phrase / petit paragraphe = un document.

Ce sont ces textes qu’on va :

nettoyer

tokeniser

encoder (One-Hot, BoW, TF-IDF).

```python
corpus = [
"La transformation numérique accélère l’adoption de l’intelligence artificielle dans de nombreux secteurs économiques."
"De jeunes entreprises innovantes développent des solutions basées sur l’analyse de données massives."
"Plusieurs centres de recherche collaborent pour faire progresser les applications de la data science."
"Le traitement automatique du langage naturel permet d’automatiser la classification de documents."
"Les systèmes de dialogue intelligents améliorent la relation entre les organisations et leurs utilisateurs."
"Les infrastructures urbaines intègrent désormais des réseaux de capteurs pour optimiser les services publics."
"Le concept de tourisme intelligent mise sur la personnalisation de l’expérience grâce aux données et à l’IA."
"Les progrès en apprentissage automatique attirent de plus en plus de chercheurs et d’ingénieurs."
]
```

Étape 2 — Le code de nettoyage et tokenisation

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string

# Télécharger les ressources nécessaires une seule fois
nltk.download('punkt')
nltk.download('stopwords')

# Charger les stopwords français
stop_words = set(stopwords.words('french'))

# Créer une nouvelle liste pour le corpus nettoyé
corpus_clean = []

for text in corpus:
    # Mettre en minuscules et tokenizer
    tokens = word_tokenize(text.lower())
    
    # Garder uniquement les mots alphabétiques qui ne sont pas des stopwords
    tokens = [t for t in tokens if t.isalpha() and t not in stop_words]
    
    # Reconstituer le texte nettoyé avec des espaces entre les mots
    corpus_clean.append(" ".join(tokens))

print("Corpus_nettoye:")
for i, doc in enumerate(corpus_clean):
    print(f"{i+1}. {doc}")
```

```python
tokens = corpus_clean[0].split()
len(tokens)          # nombre total de mots
len(set(tokens))     # nombre de mots distincts
```

Q2. Pourquoi est-il important de normaliser le texte avant la vectorisation ?

Normaliser (= mettre en forme propre) avant de transformer en vecteurs, ça sert à :

* Réduire le bruit

On enlève les choses peu informatives (stopwords, ponctuation).

Le modèle ne se concentre que sur les mots importants (intelligence, artificielle, données, recherche…).

* Limiter la dimension du vocabulaire

Sans nettoyage, on aurait énormément de "mots" différents :

Intelligence, intelligence, INTELLIGENCE → considérés comme différents.

Avec lower(), tout devient intelligence.

* Améliorer les futurs vecteurs (One-Hot, BoW, TF-IDF)

Moins de mots inutiles → matrices plus compactes, plus rapides à traiter.

Les méthodes statistiques captent mieux l’information utile.

Rendre les documents comparables entre eux

Même traitement pour tous les textes → les vecteurs sont cohérents.

# Partie B : One-Hot Encoding

```python
# =============================
# 1) Import de l’encodeur One-Hot
# =============================
from sklearn.preprocessing import MultiLabelBinarizer
# MultiLabelBinarizer permet de transformer
# une liste de mots → en vecteur binaire pour chaque document


# =============================
# 2) Tokenisation document par document
# =============================
# corpus_clean contient des chaînes "mot1 mot2 mot3..."
# doc.split() découpe chaque texte nettoyé en liste de mots
tokenized = [doc.split() for doc in corpus_clean]


# =============================
# 3) Création de l’objet One-Hot
# =============================
mlb = MultiLabelBinarizer()
# cet objet va apprendre tout le vocabulaire du corpus
# puis générer une matrice binaire


# =============================
# 4) Apprentissage + transformation
# =============================
onehot = mlb.fit_transform(tokenized)
# fit() → apprend tous les mots du vocabulaire
# transform() → transforme chaque document en vecteur binaire
# fit_transform fait les deux en une seule étape


# =============================
# 5) Affichage du vocabulaire appris
# =============================
print("Vocabulaire:", mlb.classes_)
# mlb.classes_ = liste triée de tous les mots rencontrés dans le corpus



# =============================
# 6) Affichage de la matrice One-Hot
# =============================
print("Matrice One-Hot :\n", onehot)
# Chaque ligne = un document
# Chaque colonne = un mot
# valeur 1 = présent dans le doc
# valeur 0 = absent
```

##Reponses aux Questions 

- 1 Que représente chaque ligne du tableau ?

Chaque ligne représente un document du corpus sous forme d’un vecteur binaire.

1 = le mot apparaît dans le document

0 = le mot n’apparaît pas dans le document

- 2 Quelle est la taille du vecteur ?

La taille du vecteur correspond au nombre de mots différents présents dans tout le corpus (le vocabulaire).

Dans la sortie :
 Le vocabulaire contient 59 mots
 
 Donc chaque document est représenté par un vecteur de dimension 59.

```python
print("Taille du vecteur :", len(mlb.classes_))
```

# Partie C : Bag of Words

```python
# ================================
# 1) Import de la classe CountVectorizer
# ================================
from sklearn.feature_extraction.text import CountVectorizer
# CountVectorizer transforme des documents textuels
# en une matrice de "comptage de mots"


# ================================
# 2) Instanciation du vectoriseur
# ================================
vectorizer = CountVectorizer()
# L'objet vectorizer va :
# - apprendre le vocabulaire du corpus
# - compter le nombre d'occurrences de chaque mot


# ================================
# 3) Apprentissage + transformation du corpus
# ================================
X_bow = vectorizer.fit_transform(corpus_clean)
# fit_transform réalise 2 actions :
# 1. fit()  = apprend tous les mots distincts du corpus (vocabulaire)
# 2. transform() = crée une matrice avec les comptages
#
# Résultat : X_bow = matrice (documents × mots uniques)


# ================================
# 4) Affichage du vocabulaire appris
# ================================
print("Vocabulaire :", vectorizer.get_feature_names_out())
# get_feature_names_out() renvoie la liste des mots (colonnes du modèle)


# ================================
# 5) Affichage de la matrice BoW
# ================================
print("\nMatrice BoW :\n", X_bow.toarray())
# X_bow est une structure "sparse" (optimisée)
# .toarray() la convertit en tableau normal lisible
#
# Chaque ligne = document
# Chaque colonne = mot du vocabulaire
# Chaque case = nombre d'occurrences du mot


# ================================
# 6) Visualisation propre avec pandas
# ================================
import pandas as pd

df_bow = pd.DataFrame(
    X_bow.toarray(),                         # Les données numériques
    columns=vectorizer.get_feature_names_out()  # Les mots comme colonnes
)

print(df_bow)
# Le tableau affiche les fréquences mot/par document
```

##Reponses aux questions:

* 1 One-Hot ne retient que la présence/absence (0 ou 1), tandis que Bag-of-Words compte le nombre d’occurrences des mots.

* 2 Si un mot apparaît plusieurs fois dans un document, sa valeur augmente (ex : 2, 3…).

* 3 Dans ce corpus, les mots les plus fréquents sont automatique, données et plus (fréquence 2).

# Partie D : TF-IDF

Principe

TF-IDF = Term Frequency – Inverse Document Frequency

Il combine 2 idées :

- 1 TF (Term Frequency)

    fréquence du mot dans le document
    plus le mot apparaît souvent, plus son poids augmente

- 2 IDF (Inverse Document Frequency)

    inverse de la fréquence du mot dans le corpus
    plus un mot apparaît dans tous les documents, plus il devient moins informatif

Un mot rare mais important → score élevé
Un mot très commun → score faible

* TF-IDF récompense les mots significatifs, et pénalise les mots banals.

```python
# =============================
# 1) Import du vectoriseur TF-IDF
# =============================
from sklearn.feature_extraction.text import TfidfVectorizer
# TfidfVectorizer transforme du texte en matrice pondérée TF-IDF


# =============================
# 2) Instanciation du vectoriseur
# =============================
tfidf = TfidfVectorizer()
# Par défaut il :
# - tokenize automatiquement
# - construit le vocabulaire
# - calcule TF + IDF pour chaque mot


# =============================
# 3) Apprentissage + transformation
# =============================
X_tfidf = tfidf.fit_transform(corpus_clean)
# fit_transform :
# 1. fit() = apprend les mots du corpus
# 2. transform() = calcule les scores TF-IDF
#
# X_tfidf = matrice (documents × vocabulaire)


# =============================
# 4) Afficher le vocabulaire
# =============================
print("Vocabulaire :", tfidf.get_feature_names_out())
# Liste des mots utilisés (colonnes de la matrice)


# =============================
# 5) Afficher la matrice TF-IDF brute
# =============================
print("\nMatrice TF-IDF :\n", X_tfidf.toarray())
# .toarray() rend la matrice lisible sous forme de tableau numpy
#
# Chaque valeur = poids TF-IDF du mot dans le document


# =============================
# 6) Visualisation claire avec pandas
# =============================
import pandas as pd

df_tfidf = pd.DataFrame(
    X_tfidf.toarray(),
    columns=tfidf.get_feature_names_out()
)
print(df_tfidf)
# Dataframe = tableau :
# lignes = documents
# colonnes = mots
# valeurs numériques = pertinence TF-IDF
```

##Réponses aux questions 

1 - Avec BoW, on obtient des valeurs entières qui comptent le nombre d’occurrences des mots. Avec TF-IDF, on obtient des valeurs décimales pondérées qui reflètent l’importance du mot.

2 - Les mots les plus pondérés sont "automatique", "données" et "plus", car ils apparaissent deux fois dans le document.

3 - TF-IDF est préféré à BoW car il pénalise les mots fréquents et peu informatifs et met en valeur les mots importants, ce qui améliore la recherche d’information.
