Charger et afficher un texte

```python
texte = """Le traitement automatique du langage naturel (NLP) est un
domaine␣fascinant de l’intelligence artificielle.
Il permet aux ordinateurs de comprendre et de generer du langage␣
humain."""
print(texte)
```

Tokenisation

```python
import nltk

nltk.download('punkt')       # modèle de tokenization
nltk.download('punkt_tab')   # pour les versions récentes de NLTK
```

```python
import nltk
from nltk.tokenize import sent_tokenize , word_tokenize
phrases = sent_tokenize(texte)
mots = word_tokenize(texte)
print("Phrases:", phrases)
print("\nMots:", mots)
```

Suppression des stopwords

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('french'))
mots_filtres = [mot for mot in mots
if mot.lower() not in stop_words and mot.isalpha ()]
print("Mots apres suppression des stopwords:", mots_filtres)
```

Lemmatisation et racination

Lemmatisation

```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')  # dictionnaires multilingues
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
lemmes = [lemmatizer.lemmatize(mot) for mot in mots_filtres]
print('Lemme :', lemmes)
```

Racination (stemming)

```python
from nltk.stem import SnowballStemmer
stemmer = SnowballStemmer("french")
racines = [stemmer.stem(mot) for mot in mots_filtres]
print("Racines␣:", racines)
```

Étiquetage grammatical (POS Tagging)

```python
import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
pos_tags = nltk.pos_tag(mots)
print(pos_tags)
```

Résumé final

```python
print("Texte original␣:", texte)
print("\nMots␣ filtres ␣:", mots_filtres)
print("Racines␣:", racines)
print("Lemmes␣:", lemmes)
```

1. Quelle est la différence entre stemming et lemmatisation ?
Stemming

C’est une méthode simple et mécanique qui coupe les mots en supprimant des suffixes ou préfixes.
- Elle ne comprend pas la grammaire ni le sens.
- Résultat = une racine approximative.

Exemples :

automatisation → automat

ordinateurs → ordin

comprendre → comprendr

Le stemming est rapide mais parfois imprécis.

- Lemmatisation

* Elle renvoie la forme canonique du mot (= lemme).
* Elle utilise :

le vocabulaire

la grammaire

des dictionnaires linguistiques

Exemples :

ordinateurs → ordinateur

générer → générer

permet → permettre

* La lemmatisation est plus intelligente, mais nécessite un modèle linguistique (WordNet, spaCy…).

* Résumé simple
Méthode	Résultat	Intelligence	Qualité
Stemming	Racine coupée	Aucune	Moyen
Lemmatisation	Mot normalisé	Linguistique	Bonne

2. Pourquoi supprime-t-on les stopwords ?

Les stopwords sont des mots très fréquents qui n’apportent pas d’information utile pour l’analyse.

Exemples en français :

le, la, de, est, aux, un, et…

- Leur présence parasite le sens lors du traitement automatique.

On les supprime pour :

* 1. Réduire le bruit

Ex : “le”, “de” apparaissent partout → ils ne nous apprennent rien.

* 2. Améliorer les algorithmes NLP

Moins de mots = traitement plus efficace.
* 3. Meilleure représentation des données

Un modèle TF-IDF ou un clustering sera plus fiable.

- Les stopwords ne sont généralement pas discriminants.

3. Avantages et limites de NLTK pour le français
* Avantages

Simple à utiliser

Très bon pour l’apprentissage

Bons outils de base :

tokenisation

stopwords

calculs statistiques

n-grams

Large communauté

* Parfait pour débuter ou pour des exercices académiques.

- Limites importantes (surtout en français)
- 1. Les modèles sont majoritairement conçus pour l’anglais

POS Tagging (pos_tag)

Lemmatizer (WordNet)

Named Entity Recognition

- Ils ne comprennent pas les règles du français.

- 2. Lemmatisation mauvaise en français

Ton exemple le montre :

"ordinateurs" → ordinateurs

"générer" → generer (pas de normalisation)

Aucun traitement des accents / conjugaisons

- 3. POS Tagging ↑

pos_tag() renvoie des tags anglais (NN, VB…), et se trompe souvent sur les mots français.

- 4. Stemming basique

Racines coupées grossièrement

Pas d’analyse morphologique

- Conclusion claire

- NLTK = bon pour la théorie, pas pour le français moderne.
- Pour un travail sérieux en français : spaCy ou Stanza.
